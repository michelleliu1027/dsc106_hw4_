### stars relation to finn! Network Graph

```python
finn_actors_shows = []
for show in finn_fimography['filmography']:
    if show['category'] == 'actor':
        finn_actors_shows.append(show['id'].split('/')[2])
```

```python
set(finn_actors_shows)
```

```python
stranger_things_credits = imdb._client.get_title_credits(imdb_id='tt4574334')
stars = stranger_things_credits['credits']['cast']
stars
```

```python
stranger_things_stars = {}
for star in stars:
    star_id = star['id'].split('/')[2]
    fimography = imdb._client.get_name_filmography(star_id)
    star_shows = []
    for show in fimography['filmography']:
        if (show['category'] == 'actor') or (show['category'] == 'actress'):
            star_shows.append(show['id'].split('/')[2])
    stranger_things_stars[star['name']] = star_shows
```

```python
relations_to_finn = {}
for name,info in stranger_things_stars.items():
    if len(set(info).intersection(set(finn_actors_shows))) != len(set(finn_actors_shows)):
        relations_to_finn[name] = len(set(info).intersection(set(finn_actors_shows)))
```

```python
relations_to_finn
```

```python
with open('relations_to_finn.json', 'w') as json_file:
    json.dump(relations_to_finn, json_file)
```
